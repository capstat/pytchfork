The main unicode issue I had was with using the fuzzywuzzy package.  Fuzzywuzzy was created by the developers at seatgeek.com.  It's pretty cool. Next, I wanted to see if Pitchfork's popularity has changed the way it reviews albums.  I thought that the ratings that they issued would get lower as the site's popularity increased.  But, it seems the opposite happened.  As the years progressed the number of poor album reviews decreased.

```python
import matplotlib.patches as mpatches
for year in years:
    temp = p_df[p_df.year == year]
    total = temp.shape[0]
    twos = (temp[temp['rating'] < 3].count())/total
    threes = ((temp[temp['rating'] < 4].count())/total) - twos[0]
    fours = ((temp[temp['rating'] < 5].count())/total) - threes[0]
    bar(year, twos[0], color='r', align='center')
    bar(year, threes[0], color='g', align='center', bottom=twos[0])
    bar(year, fours[0], color='b', align='center', bottom=threes[0])
title('All Pitchfork Album Reviews < 5 Rating')
ylabel('Proportion')
xlabel('Year')
axis([1998.5, 2015.5, 0, .14])
xticks(rotation='vertical')
L2 = mpatches.Patch(color='red', label='< 3')
L3 = mpatches.Patch(color='green', label='< 4')
L4 = mpatches.Patch(color='blue', label='< 5')
legend(handles=[L2,L3,L4])
show()
```

If we take a look at the higher rated albums, we can see a wave like pattern.

```python
for year in years:
    temp = p_df[p_df.year == year]
    total = temp.shape[0]
    temp = temp[temp.rating >= 7]
    sevens = ((temp[temp['rating'] < 8].count())/total).iloc[0]
    eights = ((temp[temp['rating'] < 9].count())/total).iloc[0] - sevens
    nines = ((temp[temp['rating'] >= 9].count())/total).iloc[0]
    bar(year, sevens, color='r', align='center')
    bar(year, eights, color='g', align='center', bottom=sevens)
    bar(year, nines, color='b', align='center', bottom=sevens+eights)
title('All Pitchfork Album Reviews > 7 Rating')
ylabel('Proportion')
xlabel('Year')
axis([1998.5, 2015.5, 0, 1])
xticks(rotation='vertical')
L7 = mpatches.Patch(color='red', label='> 7')
L8 = mpatches.Patch(color='green', label='> 8')
L9 = mpatches.Patch(color='blue', label='> 9')
legend(handles=[L9,L8,L7])
legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
show()
```
